```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
import yfinance as yf
import datetime
from datetime import date,timedelta

today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=730)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2

data = yf.download('GME', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
print(data.head())
```

```python
# get the number of training days:
print("trainging days =",data.shape)
```

```python
# Visualize the close price Data:
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.figure(figsize=(10, 4))
plt.title("GME's Stock Price")
plt.xlabel("Days")
plt.ylabel("Close Price USD ($)")
plt.plot(data["Close"])
plt.show()
```

```python
# get the close price:
data = data[["Close"]]
print(data.head())
```

```python
# Creating a variable to predict ‘X’ days in the future:
futureDays = 25
```

```python
#Create a new target column shifted ‘X’ units/days up:
data["Prediction"] = data[["Close"]].shift(-futureDays)
print(data.head())
print(data.tail())
```

```python
# To create a feature dataset (x) and convert into a numpy array and remove last ‘x’ rows/days:

import numpy as np
x = np.array(data.drop(["Prediction"], 1))[:-futureDays]
print(x)
```

```python
# To create a target dataset (y) and convert it to a numpy array and get all of the target values except the last ‘x’ rows days:

y = np.array(data["Prediction"])[:-futureDays]
print(y)
```

```python
# Split the data into 75% training and 25% testing

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25)
```

```python
# Creating the decision tree regressor model
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor().fit(xtrain, ytrain)

# creating the Linear Regression model
from sklearn.linear_model import LinearRegression
linear = LinearRegression().fit(xtrain, ytrain)
```

```python
# get the last ‘x’ rows/days of the feature dataset:

xfuture = data.drop(["Prediction"], 1)[:-futureDays]
xfuture = xfuture.tail(futureDays)
xfuture = np.array(xfuture)
print(xfuture)
```

```python
# To see the model tree prediction

treePrediction = tree.predict(xfuture)
print("Decision Tree prediction =",treePrediction)
```

```python
# To see the model linear regression prediction

linearPrediction = linear.predict(xfuture)
print("Linear regression Prediction =",linearPrediction)
```

```python
# Visualize decision tree predictions

predictions = treePrediction
valid = data[x.shape[0]:]
valid["Predictions"] = predictions
plt.figure(figsize=(10, 6))
plt.title("GME's Stock Price Prediction Model(Decision Tree Regressor Model)")
plt.xlabel("Days")
plt.ylabel("Close Price USD ($)")
plt.plot(data["Close"])
plt.plot(valid[["Close", "Predictions"]])
plt.legend(["Original", "Valid", "Predictions"])
plt.show()
```

```python
# Visualize the linear model predictions

predictions = linearPrediction
valid = data[x.shape[0]:]
valid["Predictions"] = predictions
plt.figure(figsize=(10, 6))
plt.title("GME's Stock Price Prediction Model(Linear Regression Model)")
plt.xlabel("Days")
plt.ylabel("Close Price USD ($)")
plt.plot(data["Close"])
plt.plot(valid[["Close", "Predictions"]])
plt.legend(["Original", "Valid", "Predictions"])
plt.show()
```
