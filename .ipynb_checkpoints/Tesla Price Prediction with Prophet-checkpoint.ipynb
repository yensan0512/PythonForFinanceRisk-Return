```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from prophet import Prophet
```

```python
import yfinance as yf
```

```python
import datetime
from datetime import date,timedelta

today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=730)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2

data = yf.download('TSLA', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
print(data.head())
```

```python
close = data['Close']
ax = close.plot(title='Tesla')
ax.set_xlabel('Date')
ax.set_ylabel('Close')
plt.show()
```

```python
# We only need two columns from this dataset (Date and Close), so let’s create a new DataFrame with only these two columns:
data["Date"] = pd.to_datetime(data["Date"], infer_datetime_format=True)
data = data[["Date", "Close"]]
```

```python
# As we are using the Facebook prophet model here for predicting the stock prices of Tesla so we need to rename the columns:
data = data.rename(columns={"Date" : "ds", "Close" : "y" })
```

```python
# prepared the dataset for the Facebook prophet model, now let’s predict the stock prices of Tesla:
model = Prophet()
model.fit(data)
predict = model.make_future_dataframe(periods=365)
forcast = model.predict(predict)
forcast[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail()
```

```python
graph = model.plot(forcast, xlabel="Date", ylabel="Price")
```
